```python
import reverse_geocoder as rg
import pandas as pd
```

```python
df = pd.read_csv("nigeria_health_facilities_osm.csv")
```

```python
# only keep rows where osm_type = node
df = df[df["osm_type"] == "node"]
```

```python
state_names = []

# get states each health facility is located in
for i, row in df.iterrows():
    X = row["X"]
    Y = row["Y"]
    state = rg.search((Y, X))[0]["admin1"]
    state_names.append(state)
    print("Finished", i)

# append to df
df["state_name"] = state_names
```

```python
df.to_csv("nigeria_health_facilities_osm_cleaned.csv")
```
